```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import PolynomialFeatures
```

```
df = pd.read_csv("http://www-bcf.usc.edu/~gareth/ISL/Auto.csv", na_values="?").dropna()

pol_d = np.arange(1, 11)
val_mse = []

regr = LinearRegression()
```

```
for p in pol_d:
    poly = PolynomialFeatures(p)
    X_poly = poly.fit_transform(df.horsepower.values.reshape(-1,1))
    X_train, X_val, y_train, y_val = train_test_split(X_poly,
                                                     df.mpg.ravel(),
                                                     test_size = 0.5,
                                                     random_state=42)
    regr.fit(X_train, y_train)
    poly_mse = mean_squared_error(y_val, regr.predict(X_val))
    val_mse.append(poly_mse)
```

```
plt.plot(pol_d,val_mse)
plt.title("Error de validacion")
plt.ylabel("Error cuadratico medio")
plt.xlabel("Grado del polinomio")
plt.show()
```

```
for i in range(0,10):
  pol_d=np.arange(1,11)
  val_mse=[]
  regr=LinearRegression()
  for p in pol_d:
    poly=PolynomialFeatures(p)
    X_poly=poly.fit_transform(df.horsepower.values.reshape(-1,1))
    X_train,X_val,y_train,y_val=train_test_split(X_poly,df.mpg.ravel(),test_size=0.5,random_state=i)
    regr.fit(X_train,y_train)
    poly_mse=mean_squared_error(y_val,regr.predict(X_val))
    val_mse.append(poly_mse)
  plt.plot(pol_d,val_mse)
plt.title("Error de validadcion")
plt.ylabel("Error cuadratico medio")
plt.xlabel("Grado de polinomio")
plt.show()
```
